# Exploratory Data Analysis (EDA)
Customer Churn Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Mock data loading since actual CSV isn't available in immediate env
# df = pd.read_csv('../data/raw/customer_churn_dataset.csv')

np.random.seed(42)
df = pd.DataFrame({
    'tenure': np.random.randint(1, 72, 1000),
    'MonthlyCharges': np.random.uniform(20, 120, 1000),
    'Churn': np.random.choice(['Yes', 'No'], 1000, p=[0.25, 0.75])
})
df.head()
```

## Class Imbalance

```python
sns.countplot(x='Churn', data=df)
plt.title('Churn Class Distribution')
plt.show()
```

## Visualizing Distributions

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df, x='tenure', hue='Churn', multiple='stack', bins=30)
plt.title('Tenure vs Churn')
plt.show()
```
